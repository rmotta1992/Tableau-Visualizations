```python
#import dependencies
import pandas as pd
```

```python
#read in csv files
april_2019 = pd.read_csv("data/201904-citibike-tripdata.csv")
may_2019 = pd.read_csv("data/201905-citibike-tripdata.csv")
```

```python
#preview data
may_2019.head()
```

```python
#create data frames
may2019_df = pd.DataFrame(may_2019)
april2019_df = pd.DataFrame(april_2019)
may2019_df.head()
```

```python
#merge data from both frames and preview
merged2019 = pd.merge(april2019_df, may2019_df, on= "start station id", how = "inner")
merged2019.head()
```

```python
#check data types for the frame
merged2019.type()
```

```python
#export cleaned csv
merged2019.to_csv(r'/data/cleaned-2019.csv', index=False)
```
